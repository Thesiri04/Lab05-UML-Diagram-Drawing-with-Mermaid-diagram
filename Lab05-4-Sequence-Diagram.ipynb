```
stateDiagram-v2
    [*] --> "place order"
    "place order" --> PROCESSING
    PROCESSING --> "authorize payment"
    "authorize payment" --> APPROVED: Approved
    "authorize payment" --> "NOT APPROVED": Not Approved
    APPROVED --> CAPTURED: Capture
    CAPTURED --> "Invoice & Transaction [Invoice: paid]": Paid
    CAPTURED --> "Invoice & Transaction [Invoice: completed]": Completed
    "Invoice & Transaction [Invoice: paid]" --> "invoice open for refund": Refund
    "Invoice & Transaction [Invoice: completed]" --> "invoice open for void or capture": Void or Capture

```
