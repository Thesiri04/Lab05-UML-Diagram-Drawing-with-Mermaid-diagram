2. เขียน code ในภาษา mermaid diagram เพื่อวาดตามไดอะแกรมนั้น
3. นำภาพที่ได้เทียบกับต้นฉบับ โดย capture เป็น png แล้วแนบลิ้งค์มาในไฟล์ ipynb นี้ (อย่าวางภาพลงไปโดยการ copy & paste เนื่องจากจะทำให้ไฟล์ ipynb มีขนาดที่โตเกินไป)


```
stateDiagram-v2
    [*] --> Idle
    Idle --> AddingClothes: ผู้ใช้ใส่ผ้า
    AddingClothes --> AddingDetergent: ผู้ใช้ใส่น้ำยาซักผ้า
    AddingDetergent --> Washing: เริ่มซักผ้า
    Washing --> Rinsing: ล้างน้ำ
    Rinsing --> Spinning: ปั่นแห้ง
    Spinning --> Completed: ซักเสร็จ
    Completed --> Idle: พร้อมสำหรับการซักครั้งต่อไป
```

```
stateDiagram-v2
    [*] --> Idle
    Idle --> CardInserted: ใส่บัตร
    CardInserted --> PINEntered: ป้อนรหัส PIN
    PINEntered --> SelectingTransaction: เลือกรายการ
    SelectingTransaction --> WithdrawCash: ถอนเงิน
    WithdrawCash --> EjectCard: คืนบัตร
    EjectCard --> Idle: พร้อมสำหรับผู้ใช้ถัดไป
```

```
stateDiagram-v2
    [*] --> Searching
    Searching --> Selecting: เลือกห้องพัก
    Selecting --> Booking: ดำเนินการจอง
    Booking --> PaymentPending: รอการชำระเงิน
    PaymentPending --> Confirmed: ยืนยันการจอง
    Confirmed --> Completed: การจองเสร็จสมบูรณ์
    Completed --> [*]
```
